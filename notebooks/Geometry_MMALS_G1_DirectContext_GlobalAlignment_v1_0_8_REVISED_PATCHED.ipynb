## Mathematical design

Let the context encoder output a raw vector $c^{\mathrm{raw}}_{s,a}\in\mathbb{R}^d$ for source image $s$ observed at angle $u_a$. The functional context is

\[
c_{s,a}=\frac{c^{\mathrm{raw}}_{s,a}}{\max(\lVert c^{\mathrm{raw}}_{s,a}\rVert_2,\varepsilon)}.
\]

The same $c_{s,a}$ is consumed by the context-bottleneck router and used by all geometry measurements.

### Chord-compatible local geometry

The normalized half-chord distance is

\[
d_c(c_{s,a},c_{s,b})=\frac{1}{2}\lVert c_{s,a}-c_{s,b}\rVert_2\in[0,1].
\]

For the controlled span $U_{\max}=120^\circ$, the target is

\[
\tau(u_a,u_b)=\sin\!\left(\frac{\pi}{2}\min\!\left(\frac{|u_a-u_b|}{U_{\max}},1\right)\right).
\]

The local matching term is

\[
\mathcal{L}_{\mathrm{context}}=
\mathbb{E}_{s}\frac{1}{|P|}\sum_{a<b}
\left(d_c(c_{s,a},c_{s,b})-\tau(u_a,u_b)\right)^2.
\]

A far-pair margin and a separate path-spread term prevent compressed trajectories:

\[
\mathcal{L}_{\mathrm{far}}=
\mathbb{E}_{s,a<b:\,|u_a-u_b|\geq U_{\mathrm{far}}}
\left[\max(0,m_{\mathrm{far}}-d_c)\right]^2,
\]

\[
\mathcal{L}_{\mathrm{spread}}=
\mathbb{E}_{s}\left[\max\left(0,v_{\min}-
\frac{1}{A}\sum_a\lVert c_{s,a}-\bar c_s\rVert_2^2\right)\right]^2.
\]

### Global fiber alignment

For sorted angles, define adjacent displacements

\[
\Delta c_{s,a}=c_{s,a+1}-c_{s,a}.
\]

Directions are aligned **for the same interval** across sources:

\[
\mathcal{L}_{\mathrm{fiber}}=
\frac{1}{A-1}\sum_a\frac{1}{S}\sum_s
\left(1-\left\langle
\widehat{\Delta c}_{s,a},\widehat{\overline{\Delta c}}_a
\right\rangle\right).
\]

A separate centroid loss applies the same chord geometry to factor-level context centroids. The total training objective is

\[
\mathcal{L}=\mathcal{L}_{\mathrm{CE}}
+\lambda_{\mathrm{anchor}}\mathcal{L}_{\mathrm{anchor}}
+\lambda_r\mathcal{L}_{\mathrm{route}}
+\lambda_c\mathcal{L}_{\mathrm{context}}
+\lambda_f\mathcal{L}_{\mathrm{far}}
+\lambda_v\mathcal{L}_{\mathrm{spread}}
+\lambda_a\mathcal{L}_{\mathrm{fiber}}
+\lambda_g\mathcal{L}_{\mathrm{centroid}}
+\lambda_h\mathcal{L}_{\mathrm{host-div}}.
\]

## 1. Numerical, scale-invariance, and anti-collapse self-tests


```python
factors=torch.tensor([-60.,-30.,0.,30.,60.])
for dim in [2,4]:
    raw=torch.randn(8,5,dim,requires_grad=True)
    routes=torch.softmax(torch.randn(8,5,4,requires_grad=True),-1)
    c_loss,c_diag=paired_context_geometry_loss(
        raw,factors,
        max_factor_span=CG["max_factor_span_degrees"],
        far_threshold_degrees=CG["far_threshold_degrees"],
        far_context_margin=CG["far_context_margin"],
        far_weight=CG["far_weight"],
        return_diagnostics=True,
    )
    spread,spread_diag=context_path_spread_loss(raw,CG["spread_floor"],return_diagnostics=True)
    fiber,fiber_diag=cross_source_fiber_alignment_loss(raw,factors,return_diagnostics=True)
    centroid=factor_centroid_geometry_loss(raw,factors)
    total=c_loss+0.05*spread+0.05*fiber+0.05*centroid+paired_route_geometry_loss(routes,factors)
    total.backward()
    assert torch.isfinite(total) and torch.isfinite(raw.grad).all()
    assert torch.allclose(paired_context_geometry_loss(raw.detach(),factors),paired_context_geometry_loss(13.0*raw.detach(),factors),atol=1e-6)

collapsed=torch.ones(8,5,4)
assert paired_context_geometry_loss(collapsed,factors)>0
assert context_path_spread_loss(collapsed,CG["spread_floor"])>0
assert torch.allclose(context_chord_target(torch.tensor([0.,60.,120.])),torch.tensor([0.,2**-0.5,1.]),atol=1e-6)
print("Context chord, scale-invariance, spread, fiber and backward gates: PASS")
```

## 3. Shared initialization and functional context boundary

Every context-bottleneck method uses

\[
c=\operatorname{normalize}(c^{\mathrm{raw}})
\]

for routing, geometry metrics, and interventions. `context_raw` is retained only for audit diagnostics. Methods with the same context dimension load exactly the same initial state.


```python
def build_model(context_dim):
    enc=SmallConvEncoder(latent_dim).to(DEVICE); enc.load_state_dict(sensory_state)
    return GeometryMMALS(
        enc,latent_dim=latent_dim,context_dim=context_dim,
        num_hosts=int(config["model"]["num_hosts"]),
        host_hidden_dim=int(config["model"]["host_hidden_dim"]),
        freeze_encoder=True,bottleneck_hidden_dim=64,
    ).to(DEVICE)

def state_hash(sd):
    h=hashlib.sha256()
    for name in sorted(sd):
        tensor=sd[name].detach().cpu().contiguous()
        h.update(name.encode()); h.update(tensor.numpy().tobytes())
    return h.hexdigest()

def forward_mode(model,x):
    z0=model.encode(x)
    context_raw,context=model.infer_context(z0)
    route_logits=model.context_bottleneck_router(context)
    route=torch.softmax(route_logits,-1)
    return model.synthesize(z0,context,route,context_raw=context_raw)

INITIAL_STATES={}; INITIAL_HASHES={}; PARAMETER_COUNTS={}
for dim in DIMS:
    torch.manual_seed(SEED+1080+dim)
    model=build_model(dim)
    INITIAL_STATES[dim]=copy.deepcopy(model.state_dict())
    INITIAL_HASHES[dim]=state_hash(INITIAL_STATES[dim])
    PARAMETER_COUNTS[dim]={
        "context_net":sum(p.numel() for p in model.context_net.parameters()),
        "context_router":sum(p.numel() for p in model.context_bottleneck_router.parameters()),
        "total_mediation_path":sum(p.numel() for p in model.context_net.parameters())+sum(p.numel() for p in model.context_bottleneck_router.parameters()),
    }
    del model
print(PARAMETER_COUNTS)
```

## 4. Equal-compute continual training ablations

The five dimension-2 regimes isolate the geometry terms:

1. no geometry;
2. route geometry only;
3. context geometry plus spread;
4. route plus context geometry;
5. route, context, spread, interval-wise fiber alignment, and centroid grounding.

The dimension-4 full model is a separate capacity probe, not a strictly parameter-matched treatment.


```python
def host_diversity(host_outputs):
    h=F.normalize(host_outputs,dim=-1)
    sim=torch.einsum("bhd,bjd->bhj",h,h)
    mask=~torch.eye(sim.shape[-1],dtype=torch.bool,device=sim.device)
    return sim[:,mask].square().mean()

@torch.no_grad()
def evaluate(model,method,stage):
    rows=[]; model.eval()
    for angle in all_eval_angles:
        total=correct=0; nll=0.0
        for x,y,_,_ in fixed_loader(angle,False,test_indices,False,batch_size=256):
            x,y=x.to(DEVICE),y.to(DEVICE)
            trace=forward_mode(model,x)
            total+=y.numel(); correct+=(trace.logits.argmax(1)==y).sum().item()
            nll+=float(F.cross_entropy(trace.logits,y,reduction="sum"))
        rows.append({"method":method,"stage":stage,"angle":angle,"angle_type":"train" if angle in train_angles else ("interpolation" if angle in interp_angles else "extrapolation"),"accuracy":correct/total,"nll":nll/total})
    return rows

def zero_like(tensor):
    return tensor.sum()*0.0

def train_variant(method,spec):
    dim=spec["context_dim"]
    model=build_model(dim); model.load_state_dict(INITIAL_STATES[dim],strict=True)
    params=[p for p in model.parameters() if p.requires_grad]
    optimizer=torch.optim.AdamW(params,lr=float(config["training"]["learning_rate"]),weight_decay=float(config["training"]["weight_decay"]))
    stage_rows=[]; evaluation_rows=[]
    for stage,current_angle in enumerate(PRIMARY_CURRICULUM):
        seen=PRIMARY_CURRICULUM[:stage+1]
        keys=["loss","current_ce","anchor_ce","route_geo","context_geo","context_match","context_far","context_spread","path_variance","fiber","fiber_resultant","centroid","div"]
        totals={key:0.0 for key in keys}; totals.update({"source_examples":0,"forward_images":0,"optimizer_steps":0})
        start=time.perf_counter(); model.train()
        for epoch in range(STAGE_EPOCHS):
            for views,y,factors,_ in multi_loader(seen,True,train_indices,True,epoch_seed(stage,epoch)):
                batch,angle_count=views.shape[:2]
                flat=views.reshape(-1,*views.shape[2:]).to(DEVICE)
                y=y.to(DEVICE); factors=factors.to(DEVICE)
                trace=forward_mode(model,flat)
                logits=trace.logits.reshape(batch,angle_count,-1)
                routes=trace.route.reshape(batch,angle_count,-1)
                contexts=trace.context.reshape(batch,angle_count,-1)
                hosts=trace.host_outputs.reshape(batch,angle_count,trace.host_outputs.shape[1],-1)
                current_ce=F.cross_entropy(logits[:,-1],y)
                anchor_ce=F.cross_entropy(logits[:,:-1].reshape(-1,logits.shape[-1]),y[:,None].expand(-1,angle_count-1).reshape(-1)) if angle_count>1 else zero_like(logits)
                route_geo=paired_route_geometry_loss(routes,factors) if spec["route_w"] else zero_like(routes)
                if spec["context_w"]:
                    context_geo,cdiag=paired_context_geometry_loss(
                        contexts,factors,
                        max_factor_span=CG["max_factor_span_degrees"],
                        far_threshold_degrees=CG["far_threshold_degrees"],
                        far_context_margin=CG["far_context_margin"],
                        far_weight=CG["far_weight"],
                        return_diagnostics=True,
                    )
                else:
                    context_geo=zero_like(contexts); cdiag={"context_match":zero_like(contexts),"context_far":zero_like(contexts)}
                if spec["spread_w"]:
                    spread,sdiag=context_path_spread_loss(contexts,CG["spread_floor"],return_diagnostics=True)
                else:
                    spread=zero_like(contexts); sdiag={"mean_path_variance":zero_like(contexts)}
                if spec["fiber_w"]:
                    fiber,fdiag=cross_source_fiber_alignment_loss(
                        contexts,factors,
                        direction_weight=FA["direction_weight"],
                        length_weight=FA["length_weight"],
                        return_diagnostics=True,
                    )
                else:
                    fiber=zero_like(contexts); fdiag={"fiber_resultant_length":zero_like(contexts)}
                centroid=factor_centroid_geometry_loss(
                    contexts,factors,
                    max_factor_span=CG["max_factor_span_degrees"],
                    far_threshold_degrees=CG["far_threshold_degrees"],
                    far_context_margin=CG["far_context_margin"],
                    far_weight=CG["far_weight"],
                ) if spec["centroid_w"] else zero_like(contexts)
                div=host_diversity(hosts[:,-1])
                loss=(
                    current_ce+float(W["retention_anchor"])*anchor_ce
                    +spec["route_w"]*route_geo
                    +spec["context_w"]*context_geo
                    +spec["spread_w"]*spread
                    +spec["fiber_w"]*fiber
                    +spec["centroid_w"]*centroid
                    +float(config["losses"]["host_functional_diversity"])*div
                )
                assert torch.isfinite(torch.stack([loss,current_ce,anchor_ce,route_geo,context_geo,spread,fiber,centroid,div])).all()
                optimizer.zero_grad(set_to_none=True); loss.backward(); torch.nn.utils.clip_grad_norm_(params,5.0); optimizer.step()
                totals["source_examples"]+=batch; totals["forward_images"]+=batch*angle_count; totals["optimizer_steps"]+=1
                values={
                    "loss":loss,"current_ce":current_ce,"anchor_ce":anchor_ce,"route_geo":route_geo,
                    "context_geo":context_geo,"context_match":cdiag["context_match"],"context_far":cdiag["context_far"],
                    "context_spread":spread,"path_variance":sdiag["mean_path_variance"],
                    "fiber":fiber,"fiber_resultant":fdiag["fiber_resultant_length"],"centroid":centroid,"div":div,
                }
                for key,value in values.items(): totals[key]+=float(value.detach())*batch
        count=max(totals["source_examples"],1)
        row={"method":method,"context_dim":dim,"stage":stage,"current_angle":current_angle,"seen_angles":json.dumps(seen),"forward_images":totals["forward_images"],"optimizer_steps":totals["optimizer_steps"],"source_examples":totals["source_examples"],"wall_seconds":time.perf_counter()-start,**{key:totals[key]/count for key in keys}}
        stage_rows.append(row); print(row)
        evaluation_rows.extend(evaluate(model,method,stage))
    return model,pd.DataFrame(stage_rows),pd.DataFrame(evaluation_rows)

models={}; stage_tables=[]; evaluation_tables=[]
for method in METHODS:
    print("\n===",method,"===")
    model,stage_table,evaluation_table=train_variant(method,METHOD_SPECS[method])
    models[method]=model; stage_tables.append(stage_table); evaluation_tables.append(evaluation_table)
stage_df=pd.concat(stage_tables,ignore_index=True)
evaluation_df=pd.concat(evaluation_tables,ignore_index=True)
```

Geometry metrics use the normalized functional context. Raw context is exported only to detect hidden scale use. C1 trained angles, C2 interpolation angles, and extrapolation remain separate.


```python
@torch.no_grad()
def collect_trace(model,method,angles,indices):
    rows=[]; model.eval()
    for views,labels,factors,source_ids in multi_loader(angles,False,indices,False):
        batch,angle_count=views.shape[:2]
        trace=forward_mode(model,views.reshape(-1,*views.shape[2:]).to(DEVICE))
        arrays={
            "z0":trace.z0.reshape(batch,angle_count,-1).cpu().numpy(),
            "context":trace.context.reshape(batch,angle_count,-1).cpu().numpy(),
            "context_raw":trace.context_raw.reshape(batch,angle_count,-1).cpu().numpy(),
            "route":trace.route.reshape(batch,angle_count,-1).cpu().numpy(),
            "z_mm":trace.z_mm.reshape(batch,angle_count,-1).cpu().numpy(),
            "logits":trace.logits.reshape(batch,angle_count,-1).cpu().numpy(),
        }
        for source in range(batch):
            for angle_index in range(angle_count):
                rows.append({
                    "method":method,"source_index":int(source_ids[source]),"label":int(labels[source]),
                    "angle":float(factors[source,angle_index]),
                    "correct":int(arrays["logits"][source,angle_index].argmax()==int(labels[source])),
                    "z0":arrays["z0"][source,angle_index],"context":arrays["context"][source,angle_index],
                    "context_raw":arrays["context_raw"][source,angle_index],"route":arrays["route"][source,angle_index],
                    "z_mm":arrays["z_mm"][source,angle_index],
                })
    return pd.DataFrame(rows)

def stack(frame,column): return np.stack(frame[column].to_numpy())

trace_df=pd.concat([collect_trace(models[m],m,all_eval_angles,test_indices) for m in METHODS],ignore_index=True)
PARTITIONS={"C1_trained":train_angles,"C2_interpolation":interp_angles,"extrapolation":extra_angles}
geometry_rows=[]; fiber_rows=[]; collapse_rows=[]
for partition,angles in PARTITIONS.items():
    part=trace_df[trace_df.angle.isin(angles)]
    for method in METHODS:
        sub=part[part.method==method]; factors=sub.angle.to_numpy(float); groups=sub.source_index.to_numpy()
        for space,metric,column in [("sensory","euclidean","z0"),("context","euclidean","context"),("route","fisher_rao","route"),("synthesis","euclidean","z_mm")]:
            grouped=grouped_geometry_scores(factors,stack(sub,column),groups,metric=metric)
            rho_mean,rho_low,rho_high=bootstrap_mean_ci(grouped["rho"],BOOTSTRAP_SAMPLES,seed=SEED)
            stress_mean,stress_low,stress_high=bootstrap_mean_ci(grouped["stress"],BOOTSTRAP_SAMPLES,seed=SEED+1)
            centroid=centroid_geometry_scores(factors,stack(sub,column),metric=metric)
            geometry_rows.append({"partition":partition,"method":method,"space":space,"source_rho_mean":rho_mean,"source_rho_ci_low":rho_low,"source_rho_ci_high":rho_high,"source_stress_mean":stress_mean,"source_stress_ci_low":stress_low,"source_stress_ci_high":stress_high,"centroid_rho":centroid["rho"],"centroid_stress":centroid["stress"]})
        fiber=fiber_alignment_scores(factors,stack(sub,"context"),groups)
        fiber_rows.append({"partition":partition,"method":method,"context_fiber_resultant_length":fiber["mean_resultant_length"],"source_count":fiber["source_count"]})
        diagnostics=context_collapse_diagnostics(factors,stack(sub,"context"),stack(sub,"context_raw"),groups)
        collapse_rows.append({"partition":partition,"method":method,**diagnostics})
geometry_df=pd.DataFrame(geometry_rows)
fiber_df=pd.DataFrame(fiber_rows)
collapse_df=pd.DataFrame(collapse_rows)
```

## 8. Paired treatment-control deltas for correlation and stress


```python
CONTRASTS=[
    ("route_geometry_effect_d2","cb_d2_route_geo","cb_d2_no_geo"),
    ("direct_context_geometry_effect_d2","cb_d2_context_geo","cb_d2_no_geo"),
    ("combined_geometry_effect_d2","cb_d2_route_context_geo","cb_d2_no_geo"),
    ("global_alignment_effect_d2","cb_d2_full_alignment","cb_d2_route_context_geo"),
    ("full_geometry_effect_d4","cb_d4_full_alignment","cb_d4_no_geo"),
    ("capacity_d4_vs_d2","cb_d4_full_alignment","cb_d2_full_alignment"),
]

SPACE_TO_TRACE_COLUMN = {
    "sensory": "z0",
    "context": "context",
    "route": "route",
    "synthesis": "z_mm",
}

def source_geometry(frame, method, angles, space):
    if space not in SPACE_TO_TRACE_COLUMN:
        raise ValueError(
            f"Unknown geometry space {space!r}. "
            f"Expected one of {sorted(SPACE_TO_TRACE_COLUMN)}."
        )

    column = SPACE_TO_TRACE_COLUMN[space]
    required = {"method", "angle", "source_index", column}
    missing = required.difference(frame.columns)
    if missing:
        raise KeyError(
            f"Trace frame is missing columns {sorted(missing)}. "
            f"Available columns: {sorted(frame.columns)}"
        )

    sub = frame[
        (frame.method == method)
        & (frame.angle.isin(angles))
    ]

    if sub.empty:
        raise ValueError(
            f"No trace rows for method={method!r}, "
            f"space={space!r}, angles={list(angles)}"
        )

    metric = "fisher_rao" if space == "route" else "euclidean"
    return grouped_geometry_scores(
        sub.angle.to_numpy(float),
        stack(sub, column),
        sub.source_index.to_numpy(),
        metric=metric,
    )

def paired_bootstrap(treatment,control,samples,seed):
    delta=np.asarray(treatment)-np.asarray(control)
    return bootstrap_mean_ci(delta,samples,seed=seed)

delta_rows=[]
for contrast,treatment,control in CONTRASTS:
    for partition,angles in {"C1_trained":train_angles,"C2_interpolation":interp_angles}.items():
        for space in ["context","route","synthesis"]:
            t=source_geometry(trace_df,treatment,angles,space); c=source_geometry(trace_df,control,angles,space)
            for metric_name in ["rho","stress"]:
                tm={group:value for group,value in zip(t["group_ids"],t[metric_name])}
                cm={group:value for group,value in zip(c["group_ids"],c[metric_name])}
                common=sorted(set(tm)&set(cm))
                mean,low,high=paired_bootstrap([tm[g] for g in common],[cm[g] for g in common],BOOTSTRAP_SAMPLES,SEED)
                delta_rows.append({"contrast":contrast,"partition":partition,"space":space,"metric":metric_name,"delta_mean":mean,"delta_ci_low":low,"delta_ci_high":high,"source_blocks":len(common)})
paired_source_deltas_df=pd.DataFrame(delta_rows)
```

## 9. Source-disjoint spherical causal probe

Because the functional context lies on a sphere, interventions are projected into the local tangent plane and then renormalized. Orthogonal controls are most informative in dimension 4; the dimension-2 sphere has only one intrinsic tangent direction.


```python
def centroid_by_angle(frame,column): return {float(angle):np.stack(group[column].to_numpy()).mean(0) for angle,group in frame.groupby("angle")}
def orthogonal_directions(tangent,count,seed):
    rng=np.random.default_rng(seed); output=[]
    while len(output)<count:
        vector=rng.normal(size=tangent.shape); vector-=np.dot(vector,tangent)*tangent; norm=np.linalg.norm(vector)
        if norm>1e-12: output.append(vector/norm)
    return output

@torch.no_grad()
def heldout_causal(method):
    model=models[method]
    fit=fit_trace[(fit_trace.method==method)&(fit_trace.angle.isin(train_angles))]
    X=stack(fit,"context").astype(float); y=fit.angle.to_numpy(float)
    X0=X-X.mean(0); y0=y-y.mean(); ridge=1e-6*max(np.trace(X0.T@X0)/max(X0.shape[1],1),1.0)
    tangent=np.linalg.solve(X0.T@X0+ridge*np.eye(X0.shape[1]),X0.T@y0); tangent/=max(np.linalg.norm(tangent),1e-12)
    route_centroids=centroid_by_angle(fit,"route")
    roots=np.stack([np.sqrt(np.clip(route_centroids[a],1e-12,None)) for a in train_angles])
    centered=np.asarray(train_angles)-np.mean(train_angles)
    route_direction=(centered[:,None]*(roots-roots.mean(0))).sum(0)/max(np.sum(centered**2),1e-12); route_direction/=max(np.linalg.norm(route_direction),1e-12)
    orthogonals=orthogonal_directions(tangent,int(protocol["causal_protocol"]["orthogonal_controls"]),SEED+81)
    x,labels,_,_=next(iter(fixed_loader(float(protocol["causal_protocol"]["probe_angle"]),False,tangent_eval_indices,False,batch_size=128)))
    x=x.to(DEVICE); base=forward_mode(model,x)
    tangent_tensor=torch.tensor(tangent,dtype=base.context.dtype,device=DEVICE)
    route_tensor=torch.tensor(route_direction,dtype=base.route.dtype,device=DEVICE)
    rows=[]
    for scale in protocol["causal_protocol"]["scales"]:
        projected=tangent_tensor-(base.context*tangent_tensor).sum(-1,keepdim=True)*base.context
        projected=F.normalize(projected,dim=-1)
        new_context=F.normalize(base.context+float(scale)*projected,dim=-1)
        new_route=torch.softmax(model.context_bottleneck_router(new_context),-1)
        delta=torch.sqrt(new_route.clamp_min(1e-12))-torch.sqrt(base.route.clamp_min(1e-12))
        signed=float((delta*route_tensor).sum(1).mean().cpu())
        control_effects=[]
        for direction in orthogonals:
            vector=torch.tensor(direction,dtype=base.context.dtype,device=DEVICE)
            local=vector-(base.context*vector).sum(-1,keepdim=True)*base.context
            local=F.normalize(local,dim=-1)
            control_context=F.normalize(base.context+float(scale)*local,dim=-1)
            route=torch.softmax(model.context_bottleneck_router(control_context),-1)
            route_delta=torch.sqrt(route.clamp_min(1e-12))-torch.sqrt(base.route.clamp_min(1e-12))
            control_effects.append(float(torch.abs((route_delta*route_tensor).sum(1)).mean().cpu()))
        orthogonal_mean=float(np.mean(control_effects)) if control_effects else float("nan")
        rows.append({"method":method,"scale":scale,"signed_route_effect":signed,"orthogonal_abs_effect_mean":orthogonal_mean,"csr":abs(signed)/max(orthogonal_mean,1e-12) if scale else 0.0})
    return pd.DataFrame(rows)

causal_methods=[method for method,spec in METHOD_SPECS.items() if spec["context_dim"]==int(protocol["causal_protocol"]["primary_dimension"])]
causal_df=pd.concat([heldout_causal(method) for method in causal_methods],ignore_index=True)
```
