# Geometry-MMALS G1 v1.0.6
## Static Route Controls and Counterbalanced Curriculum

**Scientific status:** controlled protocol implementation. Default execution is a development experiment.

**Accepted claim before execution:** C0 protocol implementation only.

**Not claimed:** C1-C6 qualification, adaptive-route superiority, a universal context manifold, backward transfer, replay-free continual learning, quantum advantage, or domain-general geometry.

v1.0.6 follows the context-mediation and curriculum-order experiment in v1.0.5. That run produced three decisive design signals:

1. removing instance-specific information from the router improved prediction, so adaptive routing must be compared with uniform and learned global static mixtures;
2. curriculum order dominated the geometry-loss effect, but the compared curricula ended on different final tasks;
3. router intervention effects require source-paired bootstrap confidence intervals, not only pooled means.

v1.0.6 therefore asks whether adaptive routing adds measurable value beyond static mixtures and whether curriculum effects persist under a counterbalanced design with the **same final task**.

### v1.0.5 seed-0 evidence motivating this revision

The v1.0.5 development archive reported:

- geometry versus paired no-geometry control: trained context-order delta **+0.0103** with a positive paired-source interval, but no prediction gain;
- retention anchor versus geometry alone: interpolation about **+0.085**, final trained accuracy about **+0.079**, and forgetting about **-0.077**;
- standard adaptive routing accuracy about **0.521**, versus about **0.617** when router `z0` was zeroed and **0.611** when both router inputs were zeroed;
- context-induced route shift only about **0.148** of the direct-`z0` route shift;
- a non-monotone curriculum reached about **0.704** interpolation accuracy, but ended on a different final angle from the monotone curricula;
- no causal candidate passed because mean CSR remained below one for the principal adaptive variants.

These are hypothesis-generating observations from one development seed, not qualified claims.

## 2. Controlled source split and deterministic loaders

The same source IDs are used at every angle. Interpolation and extrapolation angles never enter training or checkpoint selection. Loader seeds depend only on seed, curriculum position, and epoch - never on the method name.

```python
root = Path(config["data"]["root"])
base_train = MNIST(root=str(root), train=True, download=True)
base_test = MNIST(root=str(root), train=False, download=True)

rng = np.random.default_rng(SEED)
train_perm = rng.permutation(len(base_train))
test_perm = rng.permutation(len(base_test))
sensory_indices = train_perm[:SENSORY_SOURCE_LIMIT].tolist()
train_indices = train_perm[:TRAIN_SOURCE_LIMIT].tolist()
test_indices = test_perm[:TEST_SOURCE_LIMIT].tolist()
sensory_test_indices = test_perm[:SENSORY_TEST_LIMIT].tolist()


def sequence_checksum(values):
    payload = ",".join(map(str, values)).encode("utf-8")
    return hashlib.sha256(payload).hexdigest()


def make_generator(seed):
    generator = torch.Generator()
    generator.manual_seed(int(seed))
    return generator


def fixed_loader(angle, train, indices, shuffle, loader_seed=0, batch_size=128):
    ds = FixedAngleMNIST(root, angle=angle, train=train, download=True)
    ds = Subset(ds, indices)
    return DataLoader(
        ds,
        batch_size=batch_size,
        shuffle=shuffle,
        num_workers=0,
        generator=make_generator(loader_seed) if shuffle else None,
    )


def multi_loader(angles, train, indices, shuffle, loader_seed=0):
    ds = MultiAngleMNIST(
        root,
        angles=angles,
        train=train,
        indices=indices,
        download=True,
    )
    return DataLoader(
        ds,
        batch_size=SOURCE_BATCH_SIZE,
        shuffle=shuffle,
        num_workers=0,
        generator=make_generator(loader_seed) if shuffle else None,
    )


def epoch_loader_seed(stage_position, epoch):
    return SEED * 1_000_000 + stage_position * 10_000 + epoch


split_manifest = {
    "seed": SEED,
    "sensory_source_count": len(sensory_indices),
    "train_source_count": len(train_indices),
    "test_source_count": len(test_indices),
    "sensory_index_sha256": sequence_checksum(sensory_indices),
    "train_index_sha256": sequence_checksum(train_indices),
    "test_index_sha256": sequence_checksum(test_indices),
}
split_manifest
```

## 3. Pretrain, evaluate, and freeze the sensory grove

The sensory grove is a control boundary. Qualification mode stops if unrotated test accuracy is below the frozen threshold. Development mode records a warning but cannot promote C1-C6.

```python
latent_dim = int(config["model"]["latent_dim"])
encoder = SmallConvEncoder(latent_dim).to(DEVICE)
sensory_head = torch.nn.Linear(latent_dim, 10).to(DEVICE)
sensory_optimizer = torch.optim.AdamW(
    list(encoder.parameters()) + list(sensory_head.parameters()),
    lr=float(config["training"]["learning_rate"]),
    weight_decay=float(config["training"]["weight_decay"]),
)

sensory_history = []
for epoch in range(SENSORY_EPOCHS):
    loader = fixed_loader(
        0.0, True, sensory_indices, True,
        loader_seed=SEED * 1000 + epoch,
        batch_size=128,
    )
    encoder.train(); sensory_head.train()
    total = correct = 0
    loss_sum = 0.0
    for x, y, _, _ in loader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        logits = sensory_head(encoder(x))
        loss = F.cross_entropy(logits, y)
        if not torch.isfinite(loss):
            raise FloatingPointError("Non-finite sensory loss")
        sensory_optimizer.zero_grad(set_to_none=True)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(
            list(encoder.parameters()) + list(sensory_head.parameters()), 5.0
        )
        sensory_optimizer.step()
        total += y.numel()
        correct += (logits.argmax(1) == y).sum().item()
        loss_sum += float(loss.detach()) * y.numel()
    row = {"epoch": epoch, "loss": loss_sum / total, "train_accuracy": correct / total}
    sensory_history.append(row)
    print(row)


@torch.no_grad()
def evaluate_sensory(angle=0.0):
    encoder.eval(); sensory_head.eval()
    loader = fixed_loader(angle, False, sensory_test_indices, False, batch_size=256)
    total = correct = 0
    loss_sum = 0.0
    for x, y, _, _ in loader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        logits = sensory_head(encoder(x))
        loss_sum += float(F.cross_entropy(logits, y, reduction="sum"))
        total += y.numel()
        correct += (logits.argmax(1) == y).sum().item()
    return {"angle": angle, "accuracy": correct / total, "nll": loss_sum / total}


sensory_gate_result = evaluate_sensory(0.0)
print("Sensory gate:", sensory_gate_result, "threshold:", SENSORY_GATE)
if RUN_MODE == "qualification" and sensory_gate_result["accuracy"] < SENSORY_GATE:
    raise RuntimeError("Qualification blocked: frozen sensory grove below acceptance threshold")
if sensory_gate_result["accuracy"] < SENSORY_GATE:
    print("DEVELOPMENT WARNING: sensory gate not reached; no C1-C6 promotion allowed.")

sensory_state = copy.deepcopy(encoder.state_dict())
```

## 4. Shared initialization and explicit adaptive/static route policies

All methods share one immutable encoder, context network, hosts, synthesis normalization, classifier, and adaptive-router initialization.

Route policies:

- `standard`: adaptive route from `[z0, context]`;
- `context_only`: adaptive route from `[0, context]`;
- `direct_z0`: adaptive route from `[z0, 0]`;
- `uniform_static`: fixed equal route over hosts;
- `learned_static`: one trainable global route vector shared by all inputs.

Hosts always transform the real sensory representation. Only the route policy changes.

```python
def build_model():
    local_encoder = SmallConvEncoder(latent_dim).to(DEVICE)
    local_encoder.load_state_dict(sensory_state)
    return GeometryMMALS(
        local_encoder,
        latent_dim=latent_dim,
        context_dim=int(config["model"]["context_dim"]),
        num_hosts=int(config["model"]["num_hosts"]),
        host_hidden_dim=int(config["model"]["host_hidden_dim"]),
        freeze_encoder=True,
    ).to(DEVICE)


def state_dict_sha256(state_dict):
    digest = hashlib.sha256()
    for name in sorted(state_dict):
        tensor = state_dict[name].detach().cpu().contiguous()
        digest.update(name.encode("utf-8"))
        digest.update(str(tensor.dtype).encode("utf-8"))
        digest.update(str(tuple(tensor.shape)).encode("utf-8"))
        digest.update(tensor.numpy().tobytes())
    return digest.hexdigest()


def forward_with_router_mode(
    model,
    x,
    router_mode="standard",
    context_override=None,
    context_permutation=None,
):
    z0 = model.encode(x)
    context = model.context_net(z0)
    if context_override is not None:
        context_for_router = context_override
    elif context_permutation is not None:
        context_for_router = context[context_permutation]
    else:
        context_for_router = context

    if router_mode == "standard":
        route_logits = model.router(torch.cat([z0, context_for_router], dim=-1))
        route = torch.softmax(route_logits, dim=-1)
    elif router_mode == "context_only":
        route_logits = model.router(torch.cat([torch.zeros_like(z0), context_for_router], dim=-1))
        route = torch.softmax(route_logits, dim=-1)
    elif router_mode == "direct_z0":
        route_logits = model.router(torch.cat([z0, torch.zeros_like(context_for_router)], dim=-1))
        route = torch.softmax(route_logits, dim=-1)
    elif router_mode == "both_zeroed":
        route_logits = model.router(torch.cat([torch.zeros_like(z0), torch.zeros_like(context_for_router)], dim=-1))
        route = torch.softmax(route_logits, dim=-1)
    elif router_mode == "uniform_static":
        host_count = len(model.hosts)
        route = torch.full(
            (z0.shape[0], host_count),
            1.0 / host_count,
            dtype=z0.dtype,
            device=z0.device,
        )
    elif router_mode == "learned_static":
        if not hasattr(model, "static_route_logits"):
            raise AttributeError("learned_static mode requires model.static_route_logits")
        route = torch.softmax(model.static_route_logits, dim=0).expand(z0.shape[0], -1)
    else:
        raise ValueError(f"Unknown router_mode: {router_mode}")

    host_outputs = torch.stack([host(z0) for host in model.hosts], dim=1)
    z_mm = torch.einsum("bh,bhd->bd", route, host_outputs)
    z_mm = model.synthesis_norm(z_mm)
    logits = model.classifier(z_mm)
    return SimpleNamespace(
        z0=z0,
        context=context,
        route=route,
        host_outputs=host_outputs,
        z_mm=z_mm,
        logits=logits,
    )


def assert_finite_trace(trace, where):
    tensors = {
        "z0": trace.z0,
        "context": trace.context,
        "route": trace.route,
        "host_outputs": trace.host_outputs,
        "z_mm": trace.z_mm,
        "logits": trace.logits,
    }
    bad = [name for name, value in tensors.items() if not torch.isfinite(value).all()]
    if bad:
        raise FloatingPointError(f"Non-finite tensors at {where}: {bad}")


torch.manual_seed(SEED + 106)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED + 106)
base_model = build_model()
INITIAL_MMALS_STATE = copy.deepcopy(base_model.state_dict())
INITIAL_MMALS_SHA256 = state_dict_sha256(INITIAL_MMALS_STATE)
base_model.static_route_logits = nn.Parameter(
    torch.zeros(int(config["model"]["num_hosts"]), device=DEVICE)
)

with torch.no_grad():
    probe = torch.randn(4, 1, 28, 28, device=DEVICE)
    for mode in [
        "standard", "context_only", "direct_z0", "both_zeroed",
        "uniform_static", "learned_static",
    ]:
        tr = forward_with_router_mode(base_model, probe, router_mode=mode)
        assert tr.route.shape == (4, int(config["model"]["num_hosts"]))
        assert torch.allclose(tr.route.sum(1), torch.ones(4, device=DEVICE), atol=1e-5)
        assert_finite_trace(tr, f"router self-test {mode}")
    assert torch.allclose(
        forward_with_router_mode(base_model, probe, "uniform_static").route,
        forward_with_router_mode(base_model, probe, "learned_static").route,
        atol=1e-7,
    )
print("Initial MMALS SHA256:", INITIAL_MMALS_SHA256)
print("Adaptive/static route policy self-test: PASS")
del base_model
```

## 5. Core continual variants and route-policy controls

The primary geometry attribution remains:

`paired_compute_matched_no_geometry` versus `paired_geometry`.

The retention contrast remains:

`paired_geometry` versus `paired_geometry_anchor_010`.

The new adaptive-route utility contrast uses the same paired inputs and `0.10` anchor with geometry disabled:

- adaptive standard route;
- uniform static route;
- learned global static route.

This prevents a static baseline from being disadvantaged by an angle-geometry loss that a global route cannot satisfy.

```python
def host_diversity(host_outputs):
    h = F.normalize(host_outputs, dim=-1)
    sim = torch.einsum("bhd,bjd->bhj", h, h)
    mask = ~torch.eye(sim.shape[-1], dtype=torch.bool, device=sim.device)
    return sim[:, mask].square().mean()


@torch.no_grad()
def evaluate_model(model, angles, stage, method, router_mode):
    rows = []
    model.eval()
    for angle in angles:
        loader = fixed_loader(angle, False, test_indices, False, batch_size=256)
        total = correct = 0
        nll_sum = 0.0
        for x, y, _, _ in loader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            trace = forward_with_router_mode(model, x, router_mode=router_mode)
            assert_finite_trace(trace, f"eval method={method}, angle={angle}")
            nll = F.cross_entropy(trace.logits, y, reduction="sum")
            total += y.numel()
            correct += (trace.logits.argmax(1) == y).sum().item()
            nll_sum += float(nll)
        rows.append({
            "method": method,
            "stage": stage,
            "angle": angle,
            "angle_type": (
                "train" if angle in train_angles
                else "interpolation" if angle in interp_angles
                else "extrapolation"
            ),
            "accuracy": correct / total,
            "nll": nll_sum / total,
        })
    return rows


def train_variant(method, spec, curriculum):
    model = build_model()
    model.load_state_dict(INITIAL_MMALS_STATE, strict=True)
    assert state_dict_sha256(model.state_dict()) == INITIAL_MMALS_SHA256
    if spec["router_mode"] == "learned_static":
        model.static_route_logits = nn.Parameter(
            torch.zeros(int(config["model"]["num_hosts"]), device=DEVICE)
        )
        assert torch.allclose(
            torch.softmax(model.static_route_logits.detach(), dim=0),
            torch.full_like(model.static_route_logits.detach(), 1.0 / len(model.hosts)),
            atol=1e-7,
        )

    trainable = [p for p in model.parameters() if p.requires_grad]
    optimizer = torch.optim.AdamW(
        trainable,
        lr=float(config["training"]["learning_rate"]),
        weight_decay=float(config["training"]["weight_decay"]),
    )

    stage_rows = []
    evaluation_rows = []
    for stage_position, current_angle in enumerate(curriculum):
        seen_angles = curriculum[: stage_position + 1]
        totals = {
            "source_examples": 0,
            "forward_images": 0,
            "optimizer_steps": 0,
            "loss": 0.0,
            "current_ce": 0.0,
            "anchor_ce": 0.0,
            "geo": 0.0,
            "div": 0.0,
        }
        stage_start = time.perf_counter()
        model.train()
        for epoch in range(STAGE_EPOCHS):
            loader_seed = epoch_loader_seed(stage_position, epoch)
            if spec["paired"]:
                loader = multi_loader(seen_angles, True, train_indices, True, loader_seed)
            else:
                loader = fixed_loader(
                    current_angle,
                    True,
                    train_indices,
                    True,
                    loader_seed,
                    batch_size=int(spec.get("batch_size", SOURCE_BATCH_SIZE)),
                )

            for batch_id, batch in enumerate(loader):
                if spec["paired"]:
                    views, y, factors, _ = batch
                    source_count, angle_count = views.shape[:2]
                    flat_views = views.reshape(-1, *views.shape[2:]).to(DEVICE)
                    y = y.to(DEVICE)
                    factors = factors.to(DEVICE)
                    trace = forward_with_router_mode(model, flat_views, router_mode=spec["router_mode"])
                    assert_finite_trace(trace, f"{method}, stage={stage_position}, batch={batch_id}")
                    logits = trace.logits.reshape(source_count, angle_count, -1)
                    routes = trace.route.reshape(source_count, angle_count, -1)
                    hosts = trace.host_outputs.reshape(
                        source_count, angle_count, trace.host_outputs.shape[1], -1
                    )
                    current_index = angle_count - 1
                    current_ce = F.cross_entropy(logits[:, current_index], y)
                    if angle_count > 1:
                        old_logits = logits[:, :-1].reshape(-1, logits.shape[-1])
                        old_targets = y[:, None].expand(-1, angle_count - 1).reshape(-1)
                        anchor_ce = F.cross_entropy(old_logits, old_targets)
                    else:
                        anchor_ce = logits.sum() * 0.0
                    if float(spec["geometry_weight"]) > 0.0:
                        geo = paired_route_geometry_loss(
                            routes,
                            factors,
                            bandwidth=float(config["losses"]["route_bandwidth_degrees"]),
                            far_margin=float(config["losses"]["paired_route_far_margin"]),
                            far_weight=float(config["losses"]["paired_route_far_weight"]),
                            match_weight=float(config["losses"]["paired_route_match_weight"]),
                        )
                    else:
                        geo = routes.sum() * 0.0
                    div = host_diversity(hosts[:, current_index])
                    forward_images = source_count * angle_count
                else:
                    x, y, _, _ = batch
                    x, y = x.to(DEVICE), y.to(DEVICE)
                    trace = forward_with_router_mode(model, x, router_mode=spec["router_mode"])
                    assert_finite_trace(trace, f"{method}, stage={stage_position}, batch={batch_id}")
                    current_ce = F.cross_entropy(trace.logits, y)
                    anchor_ce = trace.logits.sum() * 0.0
                    geo = trace.route.sum() * 0.0
                    div = host_diversity(trace.host_outputs)
                    source_count = y.numel()
                    forward_images = source_count

                loss = (
                    float(config["losses"]["classification"]) * current_ce
                    + float(spec["anchor_weight"]) * anchor_ce
                    + float(spec["geometry_weight"]) * geo
                    + float(config["losses"]["host_functional_diversity"]) * div
                )
                if not torch.isfinite(torch.stack([current_ce, anchor_ce, geo, div, loss])).all():
                    raise FloatingPointError(f"Non-finite loss in {method}, stage={stage_position}")
                optimizer.zero_grad(set_to_none=True)
                loss.backward()
                grad_norm = torch.nn.utils.clip_grad_norm_(trainable, 5.0)
                if not torch.isfinite(torch.as_tensor(grad_norm)):
                    raise FloatingPointError("Non-finite gradient norm")
                optimizer.step()

                totals["source_examples"] += source_count
                totals["forward_images"] += forward_images
                totals["optimizer_steps"] += 1
                totals["loss"] += float(loss.detach()) * source_count
                totals["current_ce"] += float(current_ce.detach()) * source_count
                totals["anchor_ce"] += float(anchor_ce.detach()) * source_count
                totals["geo"] += float(geo.detach()) * source_count
                totals["div"] += float(div.detach()) * source_count

        count = max(totals["source_examples"], 1)
        static_route = None
        if spec["router_mode"] == "uniform_static":
            static_route = [1.0 / len(model.hosts)] * len(model.hosts)
        elif spec["router_mode"] == "learned_static":
            static_route = torch.softmax(model.static_route_logits.detach(), dim=0).cpu().tolist()
        row = {
            "method": method,
            "curriculum": json.dumps(curriculum),
            "stage": stage_position,
            "current_angle": current_angle,
            "seen_angles": json.dumps(seen_angles),
            "paired": bool(spec["paired"]),
            "router_mode": spec["router_mode"],
            "geometry_weight": float(spec["geometry_weight"]),
            "anchor_weight": float(spec["anchor_weight"]),
            "static_route": json.dumps(static_route) if static_route is not None else "",
            "source_examples": totals["source_examples"],
            "forward_images": totals["forward_images"],
            "optimizer_steps": totals["optimizer_steps"],
            "wall_seconds": time.perf_counter() - stage_start,
            "loss": totals["loss"] / count,
            "current_ce": totals["current_ce"] / count,
            "anchor_ce": totals["anchor_ce"] / count,
            "geo": totals["geo"] / count,
            "div": totals["div"] / count,
        }
        stage_rows.append(row)
        print(row)
        evaluation_rows.extend(
            evaluate_model(model, all_eval_angles, stage_position, method, spec["router_mode"])
        )
    return model, pd.DataFrame(stage_rows), pd.DataFrame(evaluation_rows)
```

```python
models = {}
stage_tables = []
evaluation_tables = []

for method in METHODS:
    print("\n=== TRAINING", method, "===")
    model, method_stage_df, method_evaluation_df = train_variant(
        method,
        METHOD_SPECS[method],
        PRIMARY_CURRICULUM,
    )
    models[method] = model
    stage_tables.append(method_stage_df)
    evaluation_tables.append(method_evaluation_df)

stage_df = pd.concat(stage_tables, ignore_index=True)
evaluation_df = pd.concat(evaluation_tables, ignore_index=True)
stage_df
```

## 7. Staged accuracy and forgetting

Forgetting is computed only after an angle has entered the curriculum. Held-out angles are never treated as learned tasks.

```python
def forgetting_table(evaluation, curriculum):
    rows = []
    final_stage = int(evaluation.stage.max())
    for method in evaluation.method.unique():
        sub = evaluation[(evaluation.method == method) & (evaluation.angle_type == "train")]
        for angle in train_angles:
            seen_stage = curriculum.index(angle)
            angle_rows = sub[(sub.angle == angle) & (sub.stage >= seen_stage)].sort_values("stage")
            best = float(angle_rows.accuracy.max())
            final = float(angle_rows[angle_rows.stage == final_stage].accuracy.iloc[0])
            rows.append({
                "method": method,
                "angle": angle,
                "best_accuracy": best,
                "final_accuracy": final,
                "forgetting": best - final,
            })
    return pd.DataFrame(rows)


forgetting_df = forgetting_table(evaluation_df, PRIMARY_CURRICULUM)
forgetting_df.groupby("method").agg(
    mean_forgetting=("forgetting", "mean"),
    final_trained_accuracy=("final_accuracy", "mean"),
)
```

## 8. Same-source traces and separated geometry partitions

C1 uses trained angles only. C2 interpolation geometry uses held-out interpolation angles only. Extrapolation is reported separately and is never folded into C1.

```python
@torch.no_grad()
def collect_paired_trace(model, method, angles, source_indices, router_mode):
    model.eval()
    rows = []
    loader = multi_loader(angles, False, source_indices, False)
    for views, labels, factors, source_ids in loader:
        source_count, angle_count = views.shape[:2]
        flat = views.reshape(-1, *views.shape[2:]).to(DEVICE)
        trace = forward_with_router_mode(model, flat, router_mode=router_mode)
        assert_finite_trace(trace, f"paired trace {method}")
        z0 = trace.z0.reshape(source_count, angle_count, -1).cpu().numpy()
        context = trace.context.reshape(source_count, angle_count, -1).cpu().numpy()
        route = trace.route.reshape(source_count, angle_count, -1).cpu().numpy()
        z_mm = trace.z_mm.reshape(source_count, angle_count, -1).cpu().numpy()
        logits = trace.logits.reshape(source_count, angle_count, -1).cpu().numpy()
        hosts = trace.host_outputs.reshape(
            source_count, angle_count, trace.host_outputs.shape[1], -1
        ).cpu().numpy()
        for b in range(source_count):
            for a in range(angle_count):
                rows.append({
                    "method": method,
                    "source_index": int(source_ids[b]),
                    "label": int(labels[b]),
                    "angle": float(factors[b, a]),
                    "prediction": int(np.argmax(logits[b, a])),
                    "correct": int(np.argmax(logits[b, a]) == int(labels[b])),
                    "z0": z0[b, a],
                    "context": context[b, a],
                    "route": route[b, a],
                    "z_mm": z_mm[b, a],
                    "hosts": hosts[b, a],
                })
    return pd.DataFrame(rows)


trace_df = pd.concat([
    collect_paired_trace(
        model,
        method,
        all_eval_angles,
        test_indices,
        METHOD_SPECS[method]["router_mode"],
    )
    for method, model in models.items()
], ignore_index=True)
print("paired trace rows:", len(trace_df))
```

```python
def stack_column(frame, name):
    return np.stack(frame[name].to_numpy())


def geometry_partition(frame, partition_name, partition_angles):
    rows = []
    part = frame[frame.angle.isin(partition_angles)]
    for method in part.method.unique():
        sub = part[part.method == method]
        factors = sub.angle.to_numpy(float)
        groups = sub.source_index.to_numpy()
        spaces = {
            "sensory": ("euclidean", stack_column(sub, "z0")),
            "context": ("euclidean", stack_column(sub, "context")),
            "route": ("fisher_rao", stack_column(sub, "route")),
            "synthesis": ("euclidean", stack_column(sub, "z_mm")),
        }
        for space, (metric_name, reps) in spaces.items():
            grouped = grouped_geometry_scores(factors, reps, groups, metric=metric_name)
            rho_mean, rho_low, rho_high = bootstrap_mean_ci(
                grouped["rho"], samples=BOOTSTRAP_SAMPLES, seed=SEED
            )
            stress_mean, stress_low, stress_high = bootstrap_mean_ci(
                grouped["stress"], samples=BOOTSTRAP_SAMPLES, seed=SEED + 1
            )
            centroid = centroid_geometry_scores(factors, reps, metric=metric_name)
            rows.append({
                "partition": partition_name,
                "method": method,
                "space": space,
                "source_rho_mean": rho_mean,
                "source_rho_ci_low": rho_low,
                "source_rho_ci_high": rho_high,
                "source_stress_mean": stress_mean,
                "source_stress_ci_low": stress_low,
                "source_stress_ci_high": stress_high,
                "centroid_rho": centroid["rho"],
                "centroid_stress": centroid["stress"],
                "source_blocks": len(grouped["rho"]),
                "status": "development_not_qualified",
            })
    return pd.DataFrame(rows)


geometry_df = pd.concat([
    geometry_partition(trace_df, "C1_trained", train_angles),
    geometry_partition(trace_df, "C2_interpolation", interp_angles),
], ignore_index=True)
geometry_df
```

## 9. Paired source-level treatment-control deltas

The geometry treatment and control share the same source IDs. v1.0.5 computes the source-level route, context, and synthesis score differences first, then bootstraps those paired differences.

```python
def source_geometry_frame(frame, method, angles, space, metric_name):
    sub = frame[(frame.method == method) & (frame.angle.isin(angles))]
    grouped = grouped_geometry_scores(
        sub.angle.to_numpy(float),
        stack_column(sub, space),
        sub.source_index.to_numpy(),
        metric=metric_name,
    )
    return pd.DataFrame({
        "source_index": grouped["group_ids"].astype(int),
        "rho": grouped["rho"],
        "stress": grouped["stress"],
    })


def paired_bootstrap_delta(values, samples, seed):
    x = np.asarray(values, dtype=float)
    x = x[np.isfinite(x)]
    if len(x) == 0:
        return float("nan"), float("nan"), float("nan")
    rng = np.random.default_rng(seed)
    boot = np.empty(samples)
    for i in range(samples):
        boot[i] = rng.choice(x, size=len(x), replace=True).mean()
    return float(x.mean()), float(np.quantile(boot, 0.025)), float(np.quantile(boot, 0.975))


PAIRED_CONTRASTS = [
    ("geometry_effect", "paired_geometry", "paired_compute_matched_no_geometry"),
    ("anchor_effect", "paired_geometry_anchor_010", "paired_geometry"),
    ("geometry_under_anchor", "paired_geometry_anchor_010", "paired_anchor_adaptive_no_geometry"),
    ("adaptive_vs_uniform_static", "paired_anchor_adaptive_no_geometry", "paired_anchor_uniform_static"),
    ("adaptive_vs_learned_static", "paired_anchor_adaptive_no_geometry", "paired_anchor_learned_static"),
]
paired_delta_rows = []
for contrast, treatment, control in PAIRED_CONTRASTS:
    for partition, angles in [("C1_trained", train_angles), ("C2_interpolation", interp_angles)]:
        for space, metric_name in [("context", "euclidean"), ("route", "fisher_rao"), ("z_mm", "euclidean")]:
            treatment_df = source_geometry_frame(trace_df, treatment, angles, space, metric_name)
            control_df = source_geometry_frame(trace_df, control, angles, space, metric_name)
            merged = treatment_df.merge(control_df, on="source_index", suffixes=("_t", "_c"))
            rho_delta = merged.rho_t - merged.rho_c
            stress_delta = merged.stress_t - merged.stress_c
            rho_mean, rho_low, rho_high = paired_bootstrap_delta(rho_delta, BOOTSTRAP_SAMPLES, SEED + 20)
            stress_mean, stress_low, stress_high = paired_bootstrap_delta(stress_delta, BOOTSTRAP_SAMPLES, SEED + 21)
            paired_delta_rows.append({
                "contrast": contrast,
                "treatment": treatment,
                "control": control,
                "partition": partition,
                "space": "synthesis" if space == "z_mm" else space,
                "paired_source_count": len(merged),
                "delta_rho_mean": rho_mean,
                "delta_rho_ci_low": rho_low,
                "delta_rho_ci_high": rho_high,
                "delta_stress_mean": stress_mean,
                "delta_stress_ci_low": stress_low,
                "delta_stress_ci_high": stress_high,
                "status": "development_not_qualified",
            })
paired_delta_df = pd.DataFrame(paired_delta_rows)
paired_delta_df
```

## 10. Held-out interpolation and source-paired prediction deltas

Route interpolation is compared with the nearest trained route. Prediction differences are paired by source image and held-out angle.

```python
def centroid_by_angle(frame, column):
    return {
        float(angle): np.stack(group[column].to_numpy()).mean(axis=0)
        for angle, group in frame.groupby("angle")
    }


def linear_neighbors(angle, trained):
    lower = max(value for value in trained if value < angle)
    upper = min(value for value in trained if value > angle)
    alpha = (angle - lower) / (upper - lower)
    return lower, upper, alpha


interpolation_rows = []
final_stage = int(evaluation_df.stage.max())
for method in METHODS:
    final_eval = evaluation_df[
        (evaluation_df.method == method)
        & (evaluation_df.stage == final_stage)
        & (evaluation_df.angle_type == "interpolation")
    ]
    sub = trace_df[trace_df.method == method]
    context_centroids = centroid_by_angle(sub, "context")
    route_centroids = centroid_by_angle(sub, "route")
    for row in final_eval.itertuples():
        angle = float(row.angle)
        lower, upper, alpha = linear_neighbors(angle, train_angles)
        c_pred = (1.0 - alpha) * context_centroids[lower] + alpha * context_centroids[upper]
        context_error = float(np.linalg.norm(context_centroids[angle] - c_pred))
        p_low = route_centroids[lower]
        p_high = route_centroids[upper]
        root_pred = (
            (1.0 - alpha) * np.sqrt(np.clip(p_low, 1e-12, None))
            + alpha * np.sqrt(np.clip(p_high, 1e-12, None))
        )
        p_pred = np.square(root_pred); p_pred = p_pred / p_pred.sum()
        actual = route_centroids[angle]
        route_error = fisher_rao_distance(actual, p_pred)
        nearest_error = min(
            fisher_rao_distance(actual, p_low),
            fisher_rao_distance(actual, p_high),
        )
        interpolation_rows.append({
            "method": method,
            "angle": angle,
            "accuracy": float(row.accuracy),
            "nll": float(row.nll),
            "context_interpolation_error": context_error,
            "route_interpolation_error": route_error,
            "nearest_route_error": nearest_error,
            "route_interpolation_beats_nearest": route_error < nearest_error,
            "status": "development_not_qualified",
        })
interpolation_df = pd.DataFrame(interpolation_rows)

prediction_delta_rows = []
for contrast, treatment, control in PAIRED_CONTRASTS:
    t = trace_df[(trace_df.method == treatment) & (trace_df.angle.isin(interp_angles))][
        ["source_index", "angle", "correct"]
    ]
    c = trace_df[(trace_df.method == control) & (trace_df.angle.isin(interp_angles))][
        ["source_index", "angle", "correct"]
    ]
    merged = t.merge(c, on=["source_index", "angle"], suffixes=("_t", "_c"))
    for angle, block in merged.groupby("angle"):
        delta = block.correct_t - block.correct_c
        mean, low, high = paired_bootstrap_delta(delta, BOOTSTRAP_SAMPLES, SEED + 30)
        prediction_delta_rows.append({
            "contrast": contrast,
            "angle": angle,
            "paired_source_count": len(block),
            "accuracy_delta_mean": mean,
            "accuracy_delta_ci_low": low,
            "accuracy_delta_ci_high": high,
            "status": "development_not_qualified",
        })
prediction_delta_df = pd.DataFrame(prediction_delta_rows)
interpolation_df, prediction_delta_df
```

## 11. Router mediation interventions with source-paired uncertainty

The anchored adaptive model is evaluated without retraining under clearly named interventions:

- `adaptive_standard`;
- `router_context_zeroed`;
- `router_context_shuffled`;
- `router_z0_zeroed`;
- `router_both_inputs_zeroed`.

Accuracy, class evidence, and route displacement are aggregated first per source image, then compared with the standard condition using paired bootstrap confidence intervals.

```python
@torch.no_grad()
def mediation_intervention_table(model, method, angles, source_indices):
    rows = []
    model.eval()
    interventions = [
        "adaptive_standard",
        "router_context_zeroed",
        "router_context_shuffled",
        "router_z0_zeroed",
        "router_both_inputs_zeroed",
    ]
    for angle in angles:
        loader = fixed_loader(angle, False, source_indices, False, batch_size=256)
        for x, y, _, source_ids in loader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            standard = forward_with_router_mode(model, x, router_mode="standard")
            standard_logp = F.log_softmax(standard.logits, -1).gather(1, y[:, None]).squeeze(1)
            permutation = torch.roll(torch.arange(len(x), device=DEVICE), shifts=1)
            traces = {
                "adaptive_standard": standard,
                "router_context_zeroed": forward_with_router_mode(
                    model, x, router_mode="standard", context_override=torch.zeros_like(standard.context)
                ),
                "router_context_shuffled": forward_with_router_mode(
                    model, x, router_mode="standard", context_permutation=permutation
                ),
                "router_z0_zeroed": forward_with_router_mode(model, x, router_mode="context_only"),
                "router_both_inputs_zeroed": forward_with_router_mode(model, x, router_mode="both_zeroed"),
            }
            for intervention in interventions:
                trace = traces[intervention]
                assert_finite_trace(trace, f"mediation {intervention}, angle={angle}")
                logp = F.log_softmax(trace.logits, -1).gather(1, y[:, None]).squeeze(1)
                route_shift = torch.linalg.vector_norm(trace.route - standard.route, dim=1)
                correct = (trace.logits.argmax(1) == y).float()
                for idx in range(len(y)):
                    rows.append({
                        "method": method,
                        "angle": angle,
                        "source_index": int(source_ids[idx]),
                        "intervention": intervention,
                        "correct": float(correct[idx].cpu()),
                        "class_log_prob_drop": float((standard_logp[idx] - logp[idx]).cpu()),
                        "route_shift": float(route_shift[idx].cpu()),
                        "status": "development_not_qualified",
                    })
    raw = pd.DataFrame(rows)
    summary = (
        raw.groupby(["method", "intervention"])
        .agg(
            accuracy=("correct", "mean"),
            mean_class_log_prob_drop=("class_log_prob_drop", "mean"),
            mean_route_shift=("route_shift", "mean"),
            source_rows=("correct", "count"),
        )
        .reset_index()
    )
    standard_accuracy = float(summary[summary.intervention == "adaptive_standard"].accuracy.iloc[0])
    summary["accuracy_drop_vs_standard"] = standard_accuracy - summary.accuracy

    source_means = (
        raw.groupby(["source_index", "intervention"])
        .agg(
            accuracy=("correct", "mean"),
            class_log_prob_drop=("class_log_prob_drop", "mean"),
            route_shift=("route_shift", "mean"),
        )
        .reset_index()
    )
    standard_source = source_means[source_means.intervention == "adaptive_standard"][
        ["source_index", "accuracy"]
    ].rename(columns={"accuracy": "standard_accuracy"})
    paired_rows = []
    for intervention in interventions[1:]:
        block = source_means[source_means.intervention == intervention].merge(
            standard_source, on="source_index", how="inner"
        )
        metrics = {
            "accuracy_delta": block.accuracy - block.standard_accuracy,
            "class_log_prob_drop": block.class_log_prob_drop,
            "route_shift": block.route_shift,
        }
        row = {
            "method": method,
            "intervention": intervention,
            "paired_source_count": len(block),
            "status": "development_not_qualified",
        }
        for metric_name, values in metrics.items():
            mean, low, high = paired_bootstrap_delta(
                values.to_numpy(float), BOOTSTRAP_SAMPLES, SEED + 60 + len(paired_rows)
            )
            row[f"{metric_name}_mean"] = mean
            row[f"{metric_name}_ci_low"] = low
            row[f"{metric_name}_ci_high"] = high
        paired_rows.append(row)
    paired = pd.DataFrame(paired_rows)

    context_shift = summary[
        summary.intervention.isin(["router_context_zeroed", "router_context_shuffled"])
    ].mean_route_shift.mean()
    z0_shift = float(summary[summary.intervention == "router_z0_zeroed"].mean_route_shift.iloc[0])
    dependency = pd.DataFrame([{
        "method": method,
        "context_route_shift_mean": float(context_shift),
        "z0_route_shift": z0_shift,
        "context_to_z0_shift_ratio": float(context_shift / max(z0_shift, 1e-12)),
        "status": "development_not_qualified",
    }])
    return raw, summary, paired, dependency


(
    mediation_raw_df,
    mediation_summary_df,
    mediation_paired_delta_df,
    mediation_dependency_df,
) = mediation_intervention_table(
    models["paired_geometry_anchor_010"],
    "paired_geometry_anchor_010",
    train_angles + interp_angles,
    test_indices,
)
mediation_summary_df, mediation_paired_delta_df, mediation_dependency_df
```

## 13. Signed causal direction gate for adaptive policies only

Static routes do not respond to context interventions and are therefore excluded from C4. Adaptive candidate evidence still requires specificity, correct orientation, monotonic magnitude, identity preservation, source-level uncertainty, and multi-seed replication.

```python
def orthogonal_directions(tangent, count, seed):
    rng = np.random.default_rng(seed)
    directions = []
    while len(directions) < count:
        vector = rng.normal(size=tangent.shape)
        vector = vector - np.dot(vector, tangent) * tangent
        norm = np.linalg.norm(vector)
        if norm > 1e-12:
            directions.append(vector / norm)
    return directions


@torch.no_grad()
def signed_causal_probe(model, method, trace, router_mode, probe_angle=15.0, scales=(-2, -1, 0, 1, 2)):
    trained = trace[trace.angle.isin(train_angles)]
    contexts = stack_column(trained, "context").astype(np.float64)
    angles = trained.angle.to_numpy(float)
    x = contexts - contexts.mean(0, keepdims=True)
    y = angles - angles.mean()
    ridge = 1e-6 * max(np.trace(x.T @ x) / max(x.shape[1], 1), 1.0)
    tangent = np.linalg.solve(x.T @ x + ridge * np.eye(x.shape[1]), x.T @ y)
    tangent = tangent / max(np.linalg.norm(tangent), 1e-12)

    route_centroids = centroid_by_angle(trained, "route")
    route_matrix = np.stack([
        np.sqrt(np.clip(route_centroids[angle], 1e-12, None))
        for angle in train_angles
    ])
    centered_angles = np.asarray(train_angles) - np.mean(train_angles)
    route_centered = route_matrix - route_matrix.mean(0, keepdims=True)
    route_direction = (
        centered_angles[:, None] * route_centered
    ).sum(axis=0) / max(np.sum(centered_angles ** 2), 1e-12)
    route_direction = route_direction / max(np.linalg.norm(route_direction), 1e-12)

    orthogonals = orthogonal_directions(tangent, count=8, seed=SEED + 9)
    loader = fixed_loader(probe_angle, False, test_indices, False, batch_size=128)
    x_batch, y_batch, _, _ = next(iter(loader))
    x_batch, y_batch = x_batch.to(DEVICE), y_batch.to(DEVICE)
    base = forward_with_router_mode(model, x_batch, router_mode=router_mode)
    base_logp = F.log_softmax(base.logits, -1).gather(1, y_batch[:, None]).squeeze(1)
    tangent_tensor = torch.tensor(tangent, dtype=base.context.dtype, device=DEVICE)
    route_dir_tensor = torch.tensor(route_direction, dtype=base.route.dtype, device=DEVICE)

    rows = []
    for scale in scales:
        new_context = base.context + float(scale) * tangent_tensor
        new_trace = forward_with_router_mode(
            model,
            x_batch,
            router_mode=router_mode,
            context_override=new_context,
        )
        delta_root = torch.sqrt(new_trace.route.clamp_min(1e-12)) - torch.sqrt(base.route.clamp_min(1e-12))
        signed_effect = float((delta_root * route_dir_tensor).sum(dim=1).mean().cpu())
        new_logp = F.log_softmax(new_trace.logits, -1).gather(1, y_batch[:, None]).squeeze(1)
        identity_drop = float((base_logp - new_logp).mean().cpu())
        orth_effects = []
        for direction in orthogonals:
            direction_tensor = torch.tensor(direction, dtype=base.context.dtype, device=DEVICE)
            orth_trace = forward_with_router_mode(
                model,
                x_batch,
                router_mode=router_mode,
                context_override=base.context + float(scale) * direction_tensor,
            )
            delta_orth = torch.sqrt(orth_trace.route.clamp_min(1e-12)) - torch.sqrt(base.route.clamp_min(1e-12))
            orth_effects.append(float(torch.abs((delta_orth * route_dir_tensor).sum(dim=1)).mean().cpu()))
        orth_mean = float(np.mean(orth_effects)) if orth_effects else float("nan")
        csr = abs(signed_effect) / max(orth_mean, 1e-12) if scale != 0 else 0.0
        orientation_match = True if scale == 0 else bool(np.sign(signed_effect) == np.sign(scale))
        rows.append({
            "method": method,
            "probe_angle": probe_angle,
            "scale": scale,
            "signed_route_effect": signed_effect,
            "orthogonal_abs_effect_mean": orth_mean,
            "causal_specificity_ratio": csr,
            "orientation_match": orientation_match,
            "class_log_prob_drop": identity_drop,
            "status": "development_not_qualified",
        })
    frame = pd.DataFrame(rows)
    abs1 = frame[frame.scale.abs() == 1].signed_route_effect.abs().mean()
    abs2 = frame[frame.scale.abs() == 2].signed_route_effect.abs().mean()
    summary = pd.DataFrame([{
        "method": method,
        "mean_csr_nonzero": frame[frame.scale != 0].causal_specificity_ratio.mean(),
        "orientation_pass": bool(frame[frame.scale != 0].orientation_match.all()),
        "monotonic_magnitude_pass": bool(abs2 > abs1),
        "max_abs_class_log_prob_drop": frame.class_log_prob_drop.abs().max(),
        "candidate_causal_gate": bool(
            frame[frame.scale != 0].orientation_match.all()
            and abs2 > abs1
            and frame[frame.scale != 0].causal_specificity_ratio.mean() > 1.5
            and frame.class_log_prob_drop.abs().max() < 0.05
        ),
        "status": "development_not_qualified",
    }])
    return frame, summary


causal_raw_tables = []
causal_summary_tables = []
for method in [
    "paired_compute_matched_no_geometry",
    "paired_geometry",
    "paired_anchor_adaptive_no_geometry",
    "paired_geometry_anchor_010",
    *( ["paired_anchor_context_only", "paired_anchor_direct_z0"] if RUN_MEDIATION_CONTROLS else [] ),
]:
    raw, summary = signed_causal_probe(
        models[method],
        method,
        trace_df[trace_df.method == method],
        METHOD_SPECS[method]["router_mode"],
    )
    causal_raw_tables.append(raw)
    causal_summary_tables.append(summary)
causal_df = pd.concat(causal_raw_tables, ignore_index=True)
causal_summary_df = pd.concat(causal_summary_tables, ignore_index=True)
causal_summary_df
```

## 14. Host route-mass and functional-ablation bundle

Route mass is not treated as specialization. Functional ablation, median effect, and positive-contribution fraction are retained for the anchored standard and mediation-trained controls.

```python
@torch.no_grad()
def host_bundle(model, method, router_mode, angles, source_indices):
    rows = []
    model.eval()
    for angle in angles:
        loader = fixed_loader(angle, False, source_indices, False, batch_size=256)
        for x, y, _, source_id in loader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            trace = forward_with_router_mode(model, x, router_mode=router_mode)
            base = F.log_softmax(trace.logits, -1).gather(1, y[:, None]).squeeze(1)
            for host in range(trace.route.shape[1]):
                route = trace.route.clone(); route[:, host] = 0.0
                route = route / route.sum(1, keepdim=True).clamp_min(1e-8)
                z = torch.einsum("bh,bhd->bd", route, trace.host_outputs)
                logits = model.classifier(model.synthesis_norm(z))
                ablated = F.log_softmax(logits, -1).gather(1, y[:, None]).squeeze(1)
                impact = (base - ablated).detach().cpu().numpy()
                mass = trace.route[:, host].detach().cpu().numpy()
                for idx, value in enumerate(impact):
                    rows.append({
                        "method": method,
                        "angle": angle,
                        "source_index": int(source_id[idx]),
                        "host": host,
                        "route_mass": float(mass[idx]),
                        "ablation_impact": float(value),
                    })
    raw = pd.DataFrame(rows)
    summary = (
        raw.groupby(["method", "angle", "host"])
        .agg(
            route_mass_mean=("route_mass", "mean"),
            ablation_mean=("ablation_impact", "mean"),
            ablation_median=("ablation_impact", "median"),
            positive_fraction=("ablation_impact", lambda values: float(np.mean(values > 0))),
            count=("ablation_impact", "count"),
        )
        .reset_index()
    )
    return raw, summary


host_methods = [
    "paired_anchor_adaptive_no_geometry",
    "paired_geometry_anchor_010",
]
if RUN_MEDIATION_CONTROLS:
    host_methods += ["paired_anchor_context_only", "paired_anchor_direct_z0"]
if RUN_STATIC_ROUTE_CONTROLS:
    host_methods += ["paired_anchor_uniform_static", "paired_anchor_learned_static"]
host_raw_tables = []
host_summary_tables = []
for method in host_methods:
    raw, summary = host_bundle(
        models[method], method, METHOD_SPECS[method]["router_mode"], train_angles + interp_angles, test_indices
    )
    host_raw_tables.append(raw); host_summary_tables.append(summary)
host_raw_df = pd.concat(host_raw_tables, ignore_index=True)
host_summary_df = pd.concat(host_summary_tables, ignore_index=True)
host_summary_df.head()
```

## 15. Development summaries, adaptive-route utility, and explicit non-promotion

The route-utility table separates three questions:

1. Does adaptive routing outperform a fixed uniform mixture?
2. Does adaptive routing outperform a learned global mixture?
3. Does the geometry term improve an already anchored adaptive route?

A positive result requires source-level uncertainty and multi-seed consistency. A single development row never promotes C1-C6.

```python
def safe_geometry_value(method, space, field):
    rows = geometry_df[
        (geometry_df.partition == "C1_trained")
        & (geometry_df.method == method)
        & (geometry_df.space == space)
    ]
    return float(rows[field].iloc[0]) if len(rows) else float("nan")


def method_summary(method):
    interp = interpolation_df[interpolation_df.method == method]
    forgetting = forgetting_df[forgetting_df.method == method]
    final_stage = int(evaluation_df.stage.max())
    final = evaluation_df[(evaluation_df.method == method) & (evaluation_df.stage == final_stage)]
    causal = causal_summary_df[causal_summary_df.method == method]
    compute = compute_summary_df[compute_summary_df.method == method].iloc[0]
    return {
        "method": method,
        "router_mode": METHOD_SPECS[method]["router_mode"],
        "trained_route_source_rho": safe_geometry_value(method, "route", "source_rho_mean"),
        "trained_route_centroid_rho": safe_geometry_value(method, "route", "centroid_rho"),
        "trained_context_source_rho": safe_geometry_value(method, "context", "source_rho_mean"),
        "mean_interpolation_accuracy": interp.accuracy.mean(),
        "mean_forgetting": forgetting.forgetting.mean(),
        "final_trained_accuracy": final[final.angle_type == "train"].accuracy.mean(),
        "mean_csr": float(causal.mean_csr_nonzero.iloc[0]) if len(causal) else float("nan"),
        "orientation_pass": bool(causal.orientation_pass.iloc[0]) if len(causal) else False,
        "candidate_causal_gate": bool(causal.candidate_causal_gate.iloc[0]) if len(causal) else False,
        "total_forward_images": int(compute.total_forward_images),
        "total_optimizer_steps": int(compute.total_optimizer_steps),
        "qualification_status": "not_qualified",
    }


method_summary_df = pd.DataFrame([method_summary(method) for method in METHODS])


def summary_metric(method, metric_name):
    row = method_summary_df[method_summary_df.method == method]
    return float(row[metric_name].iloc[0])


ROUTE_UTILITY_CONTRASTS = [
    (
        "adaptive_vs_uniform_static",
        "paired_anchor_adaptive_no_geometry",
        "paired_anchor_uniform_static",
    ),
    (
        "adaptive_vs_learned_static",
        "paired_anchor_adaptive_no_geometry",
        "paired_anchor_learned_static",
    ),
    (
        "geometry_effect_under_anchor",
        "paired_geometry_anchor_010",
        "paired_anchor_adaptive_no_geometry",
    ),
    (
        "context_only_vs_standard_geometry_anchor",
        "paired_anchor_context_only",
        "paired_geometry_anchor_010",
    ),
    (
        "direct_z0_vs_standard_geometry_anchor",
        "paired_anchor_direct_z0",
        "paired_geometry_anchor_010",
    ),
]
route_utility_rows = []
for contrast, treatment, control in ROUTE_UTILITY_CONTRASTS:
    if treatment not in METHODS or control not in METHODS:
        continue
    route_utility_rows.append({
        "contrast": contrast,
        "treatment": treatment,
        "control": control,
        "delta_final_trained_accuracy": summary_metric(treatment, "final_trained_accuracy") - summary_metric(control, "final_trained_accuracy"),
        "delta_interpolation_accuracy": summary_metric(treatment, "mean_interpolation_accuracy") - summary_metric(control, "mean_interpolation_accuracy"),
        "delta_forgetting": summary_metric(treatment, "mean_forgetting") - summary_metric(control, "mean_forgetting"),
        "delta_trained_route_source_rho": summary_metric(treatment, "trained_route_source_rho") - summary_metric(control, "trained_route_source_rho"),
        "delta_forward_images": summary_metric(treatment, "total_forward_images") - summary_metric(control, "total_forward_images"),
        "status": "development_not_qualified",
    })
route_utility_df = pd.DataFrame(route_utility_rows)

gate_summary_df = pd.DataFrame([
    {"gate": "C0_protocol_integrity", "status": "candidate_pass_after_successful_execution", "requirement": "all structural, finite-value, budget, static-route, curriculum-balance, split, and export assertions pass"},
    {"gate": "C1_trained_geometry", "status": "not_qualified", "requirement": "frozen paired-delta thresholds and multi-seed lower confidence bounds"},
    {"gate": "C2_interpolation", "status": "not_qualified", "requirement": "source-paired accuracy and route-interpolation gains across seeds"},
    {"gate": "adaptive_route_utility", "status": "not_qualified", "requirement": "adaptive route outperforms uniform and learned-static mixtures at equal compute"},
    {"gate": "curriculum_robustness", "status": "not_qualified", "requirement": "bounded variation across same-final-task counterbalanced orders"},
    {"gate": "C3_host_specialization", "status": "not_qualified", "requirement": "locality, functional ablation, and cross-seed role matching"},
    {"gate": "C4_causal_geometry", "status": "not_qualified", "requirement": "CSR, orientation, monotonicity, identity preservation, and source-bootstrap CI"},
    {"gate": "C5_memory_transport", "status": "not_implemented", "requirement": "reconstructive and synthetic memory transport experiments"},
    {"gate": "C6_operational_utility", "status": "not_qualified", "requirement": "accuracy-retention-cost benefit over hardened controls"},
])

method_summary_df, route_utility_df, gate_summary_df
```
